```python
#Q2(Ans) what is scatter plot ?
A scatter plot is a visual representation of how two variables relate to each other. 
You can use scatter plots to explore the relationship between two variables
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
colour=np.random.rand(50)
plt.scatter(x,y,c=colour)
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title("scatter graph")
```

```python
Q3(Ans)
The subplot() Function
The subplot() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,3,1)
plt.plot(x,y) 
plt.show()

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,3,2)
plt.plot(x1,y1)
plt.show()


x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,3,2)
plt.plot(x2,y2)
plt.show()

x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,3,4)
plt.plot(x3,y3)
plt.show()
```

```python
Q4(Ans)
Bar charts are one of the many techniques used to present data in a visual form so that the reader may readily recognize patterns or trends. 
Bar charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```
